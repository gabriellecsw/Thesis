````python
ds_nc_sig = ds_nc[(ds_nc.variants=='CTCATGGC')|(ds_nc.variants=='GTCATGTG')|(ds_nc.variants=='ATCATGGT')|(ds_nc.variants=='TTCATGAT')].copy()

ds_nc_sig = ds_nc_sig.melt(id_vars='variants',var_name='rep_pop',value_name='bc_normalized_count')
ds_nc_sig['rep'] = ds_nc_sig['rep_pop'].str.split('_').str[0]
ds_nc_sig['pop'] = ds_nc_sig['rep_pop'].str.split('_').str[1]

ds_vsd_cc_sig = ds_vsd_cc[(ds_vsd_cc.variants=='CTCATGGC')|(ds_vsd_cc.variants=='GTCATGTG')|(ds_vsd_cc.variants=='ATCATGGT')|(ds_vsd_cc.variants=='TTCATGAT')].copy()

ds_vsd_cc_sig = ds_vsd_cc_sig.melt(id_vars='variants',var_name='rep_pop',value_name='bc_normalized_count')
ds_vsd_cc_sig['rep'] = ds_vsd_cc_sig['rep_pop'].str.split('_').str[0]
ds_vsd_cc_sig['pop'] = ds_vsd_cc_sig['rep_pop'].str.split('_').str[1]

ds_vsd_sig = ds_vsd[(ds_vsd.variants=='CTCATGGC')|(ds_vsd.variants=='GTCATGTG')|(ds_vsd.variants=='ATCATGGT')|(ds_vsd.variants=='TTCATGAT')].copy()

ds_vsd_sig = ds_vsd_sig.melt(id_vars='variants',var_name='rep_pop',value_name='bc_normalized_count')
ds_vsd_sig['rep'] = ds_vsd_sig['rep_pop'].str.split('_').str[0]
ds_vsd_sig['pop'] = ds_vsd_sig['rep_pop'].str.split('_').str[1]
````

````python
variant_order = ds_nc_sig["pop"].unique().tolist()

g = sns.FacetGrid(ds_nc_sig, col="variants", col_wrap=2, hue='rep', margin_titles=True, sharey=False, height=5)
g.map(sns.stripplot, "pop", "bc_normalized_count", order=variant_order, dodge=True, jitter=True)

g.set_axis_labels("Population", "Normalized Counts")
for ax in g.axes.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(90)
g.add_legend(title='Biological Replicate')
g.fig.suptitle("Stripplot Before Batch Effect Correction (Norm Counts from DeSeq2)", fontsize=16)
plt.tight_layout()
plt.show()
````

````python
threshold = -np.log10(0.1)

# Create FacetGrid
g = sns.FacetGrid(ds[ds.comparison!='p4_vs_p5'], col="comparison", col_wrap=3, height=2, sharex=True, sharey=True)

# Plot the points
g.map_dataframe(
    sns.scatterplot,
    x="log2FoldChange",
    y="minus_log10_pvalue",
    color="black",
    alpha=0.5
)

# Annotate significant and CTCATGGC-containing points
def label_points(data, color, **kwargs):
    ax = plt.gca()
    for _, row in data.iterrows():
        x = row["log2FoldChange"]
        y = row["minus_log10_pvalue"]
        variant = row["variants"]
        if "CTCATGGC" in variant:
            ax.text(x, y, variant, fontsize=6, color="blue", ha="center", va="bottom")
        elif y > threshold:
            ax.text(x, y, variant, fontsize=6, color="darkgreen", ha="center", va="bottom")

g.map_dataframe(label_points)

# Add threshold lines
for ax in g.axes.flat:
    ax.axhline(y=threshold, color='red', linestyle='--')
    ax.axvline(x=0, color='gray', linestyle='--')

# Labels and layout
g.set_axis_labels("Log2 Fold Change", "-log10(p-value)")
g.set_titles(col_template="{col_name}")
plt.tight_layout()
plt.savefig("volcano_plot_facet.png", dpi=300, bbox_inches='tight')
plt.show()
````

````python
import regex as re

word = "apple"

# Pattern allowing at most 1 error of any type
pattern_e1 = r"apple{d<=1}"

test_strings_e1 = [
    "apple",    # 0 errors
    "aple",     # 1 deletion (missing 'p')
    "appple",   # 1 insertion (extra 'p')
    "apply",    # 1 substitution ('e' replaced by 'y')
    "apples",   # 1 insertion ('s')
    "palpe",    # 2 errors (e.g., 'a' substituted for 'p', 'p' substituted for 'a')
    "appl",     # 1 deletion ('e')
    "appplee",  # 2 errors (e.g., 1 insertion and 1 substitution, or 2 insertions)
    "applyy",   # 2 errors (1 substitution, 1 insertion)
    "orange"    # Many errors
]

print(f"--- Testing '{{d<=1}}' for '{word}' ---")
for text in test_strings_e1:
    match = re.match(pattern_e1, text)
    if match:
        print(f"'{text}': MATCHES (Fuzzy match info: {match.fuzzy_counts})")
    else:
        print(f"'{text}': DOES NOT MATCH")

# Example with more errors allowed: {e<=2}
pattern_e2 = r"apple{d<=2}"

test_strings_e2 = [
    "apple",     # 0 errors
    "apl",       # 2 deletions ('p', 'e')
    "applee",    # 1 insertion ('e')
    "applys",    # 1 substitution, 1 insertion (2 errors total)
    "aply",      # 1 deletion, 1 substitution (2 errors total)
    "palpe",     # 2 errors (e.g., p/a substitution, a/l deletion + l/p insertion is messy to count exactly, but it will be within 2)
    "ore"        # Many errors
]

print(f"\n--- Testing '{{d<=2}}' for '{word}' ---")
for text in test_strings_e2:
    match = re.match(pattern_e2, text)
    if match:
        print(f"'{text}': MATCHES (Fuzzy match info: {match.fuzzy_counts})")
    else:
        print(f"'{text}': DOES NOT MATCH")
````
