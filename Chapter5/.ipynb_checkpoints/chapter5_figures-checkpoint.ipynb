```python
br3_ctrl = pd.read_csv('measurements/110924_TFN_130924_CYT_measurements_ctrl_2026.tsv', sep ='\t')

# Treatment (tr)
br3_ctrl['tr'] = np.where(br3_ctrl['Image'].str.contains(r'_[AC]_'), 1, 2)

# Condition (protein)
br3_ctrl['treatment'] = 'control'

br3_ctrl.insert(0, 'rep', 3)
br3_ctrl['Image'] = br3_ctrl['Image'].str.extract(r'^([^_]+)')

br3_p1 = pd.read_csv('measurements/110924_TFN_130924_CYT_measurements_p1_2026.tsv', sep ='\t')

#Treatment (tr)
br3_p1['tr'] = np.where(br3_p1['Image'].str.contains(r'_[AC]_'), 1, 2)

# Condition (protein)
br3_p1['treatment'] = np.where(br3_p1['Image'].str.contains(r'_[AB]_'), 'eif4a3', 'magoh')

br3_p1.insert(0, 'rep', 3)
br3_p1['Image'] = br3_p1['Image'].str.extract(r'^([^_]+)')
```

```python
br4_ctrl = pd.read_csv('measurements/180924_TFN_200924_CYT_measurement_ctrl_2026.tsv', sep ='\t')

# Treatment (tr)
br4_ctrl['tr'] = np.where(br4_ctrl['Image'].str.contains(r'_[AC]_'), 1, 2)

# Condition (protein)
br4_ctrl['treatment'] = 'control'

br4_ctrl.insert(0, 'rep', 4)
br4_ctrl['Image'] = br4_ctrl['Image'].str.extract(r'^([^_]+)')

br4_p1 = pd.read_csv('measurements/180924_TFN_200924_CYT_measurements_p1_2026.tsv', sep ='\t')

#Treatment (tr)
br4_p1['tr'] = np.where(br4_p1['Image'].str.contains(r'_[AC]_'), 1, 2)

# Condition (protein)
br4_p1['treatment'] = np.where(br4_p1['Image'].str.contains(r'_[AB]_'), 'eif4a3', 'magoh')

br4_p1.insert(0, 'rep', 4)
br4_p1['Image'] = br4_p1['Image'].str.extract(r'^([^_]+)')
```

```python
# Exporting for supplementary file for thesis
supp_br2 = br2_p1[br2_p1.treatment=='magoh']
supp_br2 = supp_br2[['Image', 'Classification', 'Nucleus: Area', 'Nucleus: RFP mean', 'Nucleus: RFP sum', 'Nucleus: DAPI mean', 'Nucleus: DAPI sum']]
supp_br2.to_csv('measurements/cytation_rep1_magoh_overexpression_measurements.tsv', sep='\t', index=False)

supp_br3 = br3_p1[br3_p1.treatment=='magoh']
supp_br3 = supp_br3[['Image', 'Classification', 'Nucleus: Area', 'Nucleus: RFP mean', 'Nucleus: RFP sum', 'Nucleus: DAPI mean', 'Nucleus: DAPI sum']]
supp_br3.to_csv('measurements/cytation_rep2_magoh_overexpression_measurements.tsv', sep='\t', index=False)

supp_br4 = br4_p1[br4_p1.treatment=='magoh']
supp_br4 = supp_br4[['Image', 'Classification', 'Nucleus: Area', 'Nucleus: RFP mean', 'Nucleus: RFP sum', 'Nucleus: DAPI mean', 'Nucleus: DAPI sum']]
supp_br4.to_csv('measurements/cytation_rep3_magoh_overexpression_measurements.tsv', sep='\t', index=False)
```

```python
df = pd.concat([br1_ctrl, br2_ctrl, br3_ctrl, br4_ctrl,
                br1_p1, br2_p1, br3_p1, br4_p1]).reset_index(drop=True)

df = df.rename(columns={'Image':'samples', 'Nucleus: Area':'area', 'Nucleus: RFP mean': 'rfp_mean','Nucleus: DAPI mean':'dapi_mean'})
df = df[['rep','tr','treatment','samples','area','rfp_mean','dapi_mean']]

# To get normalised values
median_red = df[df['samples']=='Intronless'].groupby('rep')['rfp_mean'].agg('median').reset_index()
median_blue = df[df['samples']=='Intronless'].groupby('rep')['dapi_mean'].agg('median').reset_index()
median_red.columns = ['rep','rfp_med']
median_blue.columns = ['rep','dapi_med']
median = pd.merge(median_red, median_blue)

df = pd.merge(df, median)

df['rfp_norm'] = df['rfp_mean'] / df['rfp_med']
df['dapi_norm'] = df['dapi_mean'] / df['dapi_med']

# Ratio 
df["RatioBR"] = df["dapi_mean"]/df["rfp_mean"]
df["RatioRB"] = df["rfp_mean"]/df["dapi_mean"]

df["RatioBR_norm"] = df["dapi_norm"]/df["rfp_norm"]
df["RatioRB_norm"] = df["rfp_norm"]/df["dapi_norm"]

df = df[df.treatment!='eif4a3'] # only looking at MAGOH
```

## Statistical tests 

### Comparison between constructs within the same treatment

#### Aggregated

```python
var = ['uAUG','uORF', 'mAUG']
for i in var:
    g1 = df[(df.rep!=1) & (df.treatment=='control') & (df.samples==i)]['RatioBR']
    g2 = df[(df.rep!=1) & (df.treatment=='control') & (df.samples=='mORF')]['RatioBR']
    
    t_stat, t_pval = stats.ttest_ind(g1, g2)
    print(f"{i} vs mORF for RatioBR, t-test: stat = {t_stat}, p-value = {t_pval}")
```

```python
df_med = (
    df_plot
    .groupby(['treatment','samples', 'rep'], as_index=False)
    .agg({'RatioBR': 'median', 'RatioRB': 'median'})
)

var = ['uAUG','uORF', 'mAUG','mORF']
treatment = ['control','magoh']

results_br = []

for tr in treatment:
    for i in var:
        g1 = df_med[(df_med['samples'] == i) & (df_med['treatment'] == tr)]['RatioBR']
        g2 = df_med[(df_med['samples'] == 'mAUG') & (df_med['treatment'] == tr)]['RatioBR']
        
        t_stat, t_pval = stats.ttest_ind(g1, g2)
    
        results_br.append({
            'treatment': tr,
            'samples': i,
            'reference': 'Intronless',
            't_stat': t_stat,
            'p_value': t_pval,
            
        })

results_rb = []

for tr in treatment:
    for i in var:
        g1 = df_med[(df_med['samples'] == i) & (df_med['treatment'] == tr)]['RatioRB']
        g2 = df_med[(df_med['samples'] == 'uAUG') & (df_med['treatment'] == tr)]['RatioRB']
        
        t_stat, t_pval = stats.ttest_ind(g1, g2)
        
        results_rb.append({
            'treatment': tr,
            'samples': i,
            'reference': 'Intronless',
            't_stat': t_stat,
            'p_value': t_pval
        })

df_ttest_br = pd.DataFrame(results_br)
df_ttest_rb = pd.DataFrame(results_rb)
```

```python
plt.figure(figsize=(3, 3))

samples = ['uAUG','uORF','mAUG', 'mORF']

# --- RatioBR ---
sns.pointplot(
    data=df_ttest_br,
    x='samples',
    y='t_stat',
    order=samples,
    hue='treatment',
    errorbar=('ci', 91),   # or 'se'
    capsize=0.2, linestyles='none'
)
# plt.title('log2FC (RatioBR)')
plt.xlabel('')
plt.ylabel('')
# plt.savefig('facs23/facs23_plots/log2fc_ratiobr_ttest.png', dpi=300)
# plt.savefig('facs23/facs23_plots/log2fc_ratiobr_ttest.pdf')
# plt.show()

plt.figure(figsize=(3, 3))

# --- RatioRB  ---
sns.pointplot(
    data=df_ttest_rb,
    x='samples',
    y='t_stat',
    order=samples,
    hue='treatment',
    errorbar=('ci', 91),
    capsize=0.2, linestyles='none'
)
plt.xlabel('')
plt.ylabel('')
# plt.savefig('facs23/facs23_plots/log2fc_ratiorb_ttest.png', dpi=300)
# plt.savefig('facs23/facs23_plots/log2fc_ratiorb_ttest.pdf')
plt.show()
```
