# FACS 2023

## Flowkit, gating & extract the raw readings

```python
#Read in the .fcs files for 2023facs
samples_br1 = []
for i in glob("../../2024/facs/051223_facs/docs/*.fcs"):
    sample = fk.Sample(i, sample_id = i.replace(".fcs", "").replace("../../2024/facs/051223_facs/docs/Specimen_001_",""))
    samples_br1.append(sample)

samples_br2 = []
for i in glob("../../2024/facs/071223_facs/docs/*.fcs"):
    sample = fk.Sample(i, sample_id = i.replace(".fcs", "").replace("../../2024/facs/071223_facs/docs/Specimen_001_",""))
    samples_br2.append(sample)

samples_br3 = []
for i in glob("../../2024/facs/081223_facs/docs/*.fcs"):
    sample = fk.Sample(i, sample_id = i.replace(".fcs", "").replace("../../2024/facs/081223_facs/docs/Specimen_001_",""))
    samples_br3.append(sample)
print(samples_br3)
```

```python
#To show the channels available
samples_br1[1].channels
#assign a name for each channel so it is easier to refer back
fsc_a = fk.Dimension('FSC-A')
ssc_a = fk.Dimension('SSC-A')
fsc_h = fk.Dimension('FSC-H')
ssc_h = fk.Dimension('SSC-H')
ld_nir = fk.Dimension('APC-Cy7-A')
txred = fk.Dimension('PE-Texas Red-A')
dapi = fk.Dimension('DAPI-A')
```

```python
#Create an empty gating strategy and add the gates one by one
g_strat = fk.GatingStrategy()

#x:fsc_a, y:ssc_a
vertices_c = [(10000, 6000), (20000, 6000), (60000, 20000), (65000, 60000), (34000, 58000)]
c = fk.gates.PolygonGate('c', dimensions=[fsc_a, ssc_a], vertices=vertices_c)
g_strat.add_gate(c, gate_path=('root',))

#single cell, x:fsc_a, y:fsc_h
vertices_sc = [(10000, 10000),(22000, 10000),(67000, 24000),(65000,30000), (40000,25000), (28000,20000)]
sc = fk.gates.PolygonGate('sc', dimensions=[fsc_a, fsc_h], vertices=vertices_sc)
g_strat.add_gate(sc, gate_path=('root','c'))

#single cell2, x:ssc_a, y:ssc_h
vertices_sc2 = [(5000, 4000),(13000, 4000),(62000, 25000),(52000, 29000), (38000,26000), (5000,11000)]
sc2 = fk.gates.PolygonGate('sc2', dimensions=[ssc_a, ssc_h], vertices=vertices_sc2)
g_strat.add_gate(sc2, gate_path=('root','c','sc'))

#live_dead, x:ld_nir, y:fsc_h
vertices_ld = [(-100, 10000),(100, 10000),(100, 35000),(-100, 35000)]
ld = fk.gates.PolygonGate('ld', dimensions=[ld_nir, fsc_h], vertices=vertices_ld)
g_strat.add_gate(ld, gate_path=('root','c','sc','sc2'))

#upper pop, x:txred, y:dapi
vertices_upop = [(100, 500),(10000, 10000),(10000, 100000),(10, 10000),(10,1000)]
upop = fk.gates.PolygonGate('upop', dimensions=[txred, dapi], vertices=vertices_upop)
g_strat.add_gate(upop, gate_path=('root','c','sc','sc2','ld'))

session_br1 = fk.Session(gating_strategy= g_strat, fcs_samples=samples_br1)
session_br1.analyze_samples(verbose=False)

session_br2 = fk.Session(gating_strategy= g_strat, fcs_samples=samples_br2)
session_br2.analyze_samples(verbose=False)

session_br3 = fk.Session(gating_strategy= g_strat, fcs_samples=samples_br3)
session_br3.analyze_samples(verbose=False)
```

```python
logicle_xform = fk.transforms.LogicleTransform(
    param_t=262144,
    param_w=0.5,
    param_m=4.5,
    param_a=0
)
session_br1.add_transform('logicle', logicle_xform)
session_br1.get_transforms()

# session_br2.add_transform('logicle', logicle_xform)
# session_br2.get_transforms()

# session_br3.add_transform('logicle', logicle_xform)
# session_br3.get_transforms()
```

## Retrieve gated events

### Rep 1

```python
sample_list_br1 = session_br1.get_sample_ids()

gates = ['c', 'sc', 'sc2', 'ld','upop']

gate_data = {g: [] for g in gates}

for i in sample_list_br1:
    for g in gates:
        
        ge = session_br1.get_gate_events(sample_id=i, gate_name=g)
        ge["samples"] = i
        gate_data[g].append(ge)
        
for g in gates:
        globals()[f'df_br1_{g}'] = pd.concat(gate_data[g])
    
# Mapping for sample names
names = {
    "A":"uAUG","C":"uORF","B":"mAUG","D":"mORF","w":"Intronless",
    "HK LD NIR 1000":"HK1000","HK LD 500":"HK500","negative control":"NC"
}

for g in gates:
    df = globals()[f'df_br1_{g}'].reset_index(drop=True)
    df.columns = df.columns.get_level_values(0)
    
    # Rename channels
    df = df.rename(columns={'DAPI-A': 'dapi', 'PE-Texas Red-A':'txred'})
    
    # Create combined channels
    df['txred_com'] = df.iloc[:, 6].combine_first(df.iloc[:, 11])
    df['dapi_com'] = df.iloc[:, 7].combine_first(df.iloc[:, 12])
    df['APC-Cy7-A_com'] = df.iloc[:, 8].combine_first(df.iloc[:, 13])

    channels_to_transform = ['dapi_com', 'txred_com', 'APC-Cy7-A_com']

    for ch in channels_to_transform:
        df[ch + "_logicle"] = logicle_xform.apply(df[ch].values)
    
    # Calculate ratios
    df["RatioRB"] = df["txred_com"] / df["dapi_com"]
    df["RatioBR"] = df["dapi_com"] / df["txred_com"]
    
    df["RatioRB_logicle"] = df["txred_com_logicle"] / df["dapi_com_logicle"]
    df["RatioBR_logicle"] = df["dapi_com_logicle"] / df["txred_com_logicle"]
    
    # Rename samples
    df = df.rename(columns={'samples':'File'}).reset_index(drop=True)
    df["samples"] = df.File.apply(lambda x: names[x])
    
    # Save back to global variable
    globals()[f'df_br1_{g}'] = df
```

## Visualization using flowkit

```python
print(session_br1.get_sample_ids())
print(session_br2.get_sample_ids())
print(session_br3.get_sample_ids())
```

```python
reps = {
    "rep1": session_br1,
    "rep2": session_br2,
    "rep3": session_br3,
}

def rename_sample(sample_id):
    sample_name_map = {
        "A": "P1",
        "B": "P2",
        "C": "P3",
        "D": "P4",
        "w": "Intronless",
        "transfection reagent": "Mock",
        "negative control":"Untransfected"
    }

    if sample_id in sample_name_map:
        return sample_name_map[sample_id]

    if sample_id.startswith("HK"):
        return sample_id.replace("HK", "Heat_killed", 1).replace(" ", "_")

    return sample_id

# gate c
for rep_name, session in reps.items():
    for sample_id in session.get_sample_ids():
        nice_name = rename_sample(sample_id)
        p = session.plot_gate(sample_id, 'c')
        p.aspect_ratio = 1.0
        p.title.text = nice_name

        export_png(p,filename=f"facs23/facs23_plots/gating/c_{rep_name}_{nice_name}.png")

# gate sc
for rep_name, session in reps.items():
    for sample_id in session.get_sample_ids():
        nice_name = rename_sample(sample_id)
        p = session.plot_gate(sample_id, 'sc', x_min=0, x_max=80000, y_min=5000, y_max=35000)
        p.aspect_ratio = 1.0
        p.title.text = nice_name

        export_png(p,filename=f"facs23/facs23_plots/gating/sc_{rep_name}_{nice_name}.png")

# gate sc2
for rep_name, session in reps.items():
    for sample_id in session.get_sample_ids():
        nice_name = rename_sample(sample_id)
        p = session.plot_gate(sample_id, 'sc2', x_min=0, x_max=80000, y_min=0, y_max=40000)
        p.aspect_ratio = 1.0
        p.title.text = nice_name
        
        export_png(p,filename=f"facs23/facs23_plots/gating/sc2_{rep_name}_{nice_name}.png")
```

```python
# Plot with transformed data
x_label = 'APC-Cy7-A'
y_label = 'FSC-H'

x_dim = fk.Dimension(x_label, transformation_ref='logicle')
y_dim = fk.Dimension(y_label)

for rep_name, session in reps.items():
    for sample_id in session.get_sample_ids():
        nice_name = rename_sample(sample_id)
        p = session.plot_scatter(sample_id, x_dim, y_dim, gate_name='sc2')
        p.aspect_ratio = 1.0
        p.title.text = nice_name

        export_png(p,filename=f"facs23/facs23_plots/gating/ld_{rep_name}_{nice_name}.png")
```

```python
# Plot with transformed data
x_label = 'PE-Texas Red-A'
y_label = 'DAPI-A'

x_dim = fk.Dimension(x_label, transformation_ref='logicle')
y_dim = fk.Dimension(y_label, transformation_ref='logicle')

for rep_name, session in reps.items():
    for sample_id in session.get_sample_ids():
        nice_name = rename_sample(sample_id)
        p = session.plot_scatter(sample_id, x_dim, y_dim, gate_name='sc2', x_min=-0.05, x_max=0.9, y_min=-0.05, y_max=0.9)
        p.aspect_ratio = 1.0
        p.title.text = nice_name

        export_png(p,filename=f"facs23/facs23_plots/gating/final_{rep_name}_{nice_name}.png")
```
