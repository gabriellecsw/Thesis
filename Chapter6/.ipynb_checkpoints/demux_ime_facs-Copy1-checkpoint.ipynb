```python
#assign only no retrieve
def extract_barcode_matches_short_reads(df, metadata):
        extracted_dfs = []
        count_records = []
        
        # Extract barcode lists from your metadata
        print('Extracting unique barcodes from metadata.')
        front_barcodes = metadata['front_barcode'].unique()
        rear_barcodes  = metadata['rear_barcode'].unique()
            
        # Compile for speed
        front_patterns = [re.compile(bc) for bc in front_barcodes]
        rear_patterns  = [re.compile(bc) for bc in rear_barcodes]
        
        print('Sampling 50% of input to estimate barcodes positions')
        sampled_seqs = df['seq'].dropna().sample(frac=0.5, random_state=42)
        
        print('Estimating barcodes positions from sampled input sequences.')    
        front_distances = []
        rear_distances  = []
            
        for seq in sampled_seqs:
            if pd.isna(seq):
                    continue
            
            # ---- FRONT ----
            front_positions = []
            for pat in front_patterns:
                m = pat.search(seq)
                if m:
                    front_positions.append(m.start())
            if front_positions:
                front_distances.append(min(front_positions))
            
            # ---- REAR ----
            rear_positions = []
            for pat in rear_patterns:
                m = pat.search(seq)
                if m:
                    rear_positions.append(m.start())
            if rear_positions:
                rear_distances.append(min(rear_positions))
            
        # Final medians
        median_front = np.median(front_distances)
        median_rear  = np.median(rear_distances)
    
        lengths = []
    
        for i in front_barcodes:
            length = len(i)
            lengths.append(length)
        
        median_fb_length = np.median(lengths)
    
        print('Computing space range between two barcodes.')
    
        space_len = int(median_rear - median_front - median_fb_length)
    
        # If spacing between two barcodes are 0, the space range will be set to 0, if not users can specify the +/- value for the space range
        if space_len == 0:
            space_range = "0"
            print('No spaces between two barcodes.')
        else:
            space_range = f"{int(space_len - 5)},{int(space_len + 5)}"
            print(f'Space range between two barcodes is {space_range}.')

        print('Demultiplexing in progress...')
        for _, row in metadata.iterrows():
            pair_name = row['pair']
            front_barcode = row['front_barcode']
            rear_barcode = row['rear_barcode']
            
            # Build regex pattern for this barcode pair
            pattern = fr'''(?x)
            (?P<front>{front_barcode}){{e<=1}}
            (?P<space>.{{{space_range}}})
            (?P<rear>{rear_barcode}){{e<=1}}
            '''
        
            regex = re.compile(pattern)
        
            # 4. Apply pattern to sequences
            col_name = f"extracted_{pair_name}"
        
            df_copy = df.copy()
            df_copy[col_name] = df_copy['seq'].apply(
                lambda x: regex.search(x).groupdict() if regex.search(x) else None
            )
        
            # 5. Filter and expand extracted results
            df_valid = df_copy[df_copy[col_name].notna()].copy()
            extracted_cols = df_valid[col_name].apply(pd.Series)
            df_combined = pd.concat([df_valid.drop(columns=[col_name]), extracted_cols], axis=1)
        
            # 6. Add metadata info
            df_combined['pair'] = pair_name
            df_combined['front_barcode_name'] = row['front_barcode_name']
            df_combined['rear_barcode_name'] = row['rear_barcode_name']
        
            extracted_dfs.append(df_combined)
        
            # 7. Track counts
            count_records.append({
                'pair': pair_name,
                'front_barcode_name': row['front_barcode_name'],
                'rear_barcode_name': row['rear_barcode_name'],
                'front_barcode_seq': front_barcode,
                'rear_barcode_seq': rear_barcode,
                'count': len(df_combined)
            })
        
        # Final outputs
        df_result = pd.concat(extracted_dfs).reset_index(drop=True) if extracted_dfs else pd.DataFrame()
        df_counts = pd.DataFrame(count_records)
        
        return df_result, df_counts
```

```python
#assign and retrieve function
def assign_retrieve_short_reads(df, metadata, position, feature_length, space):
        extracted_dfs = []
        count_records = []
        
        # Extract barcode lists from your metadata
        print('Extracting unique barcodes from metadata.')
        front_barcodes = metadata['front_barcode'].unique()
        rear_barcodes  = metadata['rear_barcode'].unique()
            
        # Compile for speed
        front_patterns = [re.compile(bc) for bc in front_barcodes]
        rear_patterns  = [re.compile(bc) for bc in rear_barcodes]
        
        print('Sampling 50% of input to estimate barcodes positions')
        sampled_seqs = df['seq'].dropna().sample(frac=0.5, random_state=42)
        
        print('Estimating barcodes positions from sampled input sequences.')    
        front_distances = []
        rear_distances  = []
            
        for seq in sampled_seqs:
            if pd.isna(seq):
                    continue
            
            # ---- FRONT ----
            front_positions = []
            for pat in front_patterns:
                m = pat.search(seq)
                if m:
                    front_positions.append(m.start())
            if front_positions:
                front_distances.append(min(front_positions))
            
            # ---- REAR ----
            rear_positions = []
            for pat in rear_patterns:
                m = pat.search(seq)
                if m:
                    rear_positions.append(m.start())
            if rear_positions:
                rear_distances.append(min(rear_positions))
            
        # Final medians
        median_front = np.median(front_distances)
        median_rear  = np.median(rear_distances)
    
        lengths = []
    
        for i in front_barcodes:
            length = len(i)
            lengths.append(length)
        
        median_fb_length = np.median(lengths)
    
        print('Computing space range between two barcodes.')
    
        space_len = int(median_rear - median_front - median_fb_length)
    
        # If spacing between two barcodes are 0, the space range will be set to 0, if not users can specify the +/- value for the space range
        if space_len == 0:
            space_range = "0"
            print('No spaces between two barcodes.')
        else:
            space_range = f"{int(space_len - 5)},{int(space_len + 5)}"
            print(f'Space range between two barcodes is {space_range}.')

        print('Demultiplexing in progress...')
        for _, row in metadata.iterrows():
            pair_name = row['pair']
            front_barcode = row['front_barcode']
            rear_barcode = row['rear_barcode']

            choices = {'pattern_up': True,
                       'pattern_up_spaced': True,
                       'pattern_between': True,
                       'pattern_between_spaced': True,
                       'pattern_down': True,
                       'pattern_down_spaced': True}
            
            if position not in choices:
                typer.echo(f"position must be one of {list(choices.keys())}, got: {position}")
                raise typer.Exit(code=1)
            
            # Build regex pattern for each barcode pair and pattern
            if position == 'pattern_up':
                inter_space = space_range
                pattern = fr'''(?x)
                (?P<feature>.{{{feature_length}}})
                (?P<front>{front_barcode}){{e<=1}}
                (?P<inter>.{{{inter_space}}})
                (?P<rear>{rear_barcode}){{e<=1}}
                '''
                regex=re.compile(pattern)
                
            if position == 'pattern_inter':
                inter_space = space_range
                pattern = fr'''(?x)
                (?P<front>{front_barcode}){{e<=1}}
                (?P<inter>.{{{inter_space}}})
                (?P<rear>{rear_barcode}){{e<=1}}
                '''
                regex=re.compile(pattern)
            
            if position == 'pattern_down':
                inter_space = space_range
                pattern = fr'''(?x)
                (?P<front>{front_barcode}){{e<=1}}
                (?P<inter>.{{{inter_space}}})
                (?P<rear>{rear_barcode}){{e<=1}}
                (?P<feature>.{{{feature_length}}})
                '''
                regex=re.compile(pattern)
                
            if position=='pattern_up_spaced':
                up_space_range = space
                inter_space = space_range
                
                pattern = fr'''(?x)
                (?P<feature>.{{{feature_length}}})
                (?P<up_space>.{{{up_space_range}}})
                (?P<front>{front_barcode}){{e<=1}}
                (?P<inter>.{{{inter_space}}})
                (?P<rear>{rear_barcode}){{e<=1}}
                '''
                regex=re.compile(pattern)
                
            if position == 'pattern_inter_spaced':
                inter_space = int(median_rear - median_front +  median_fb_length + space + feature_length)
                
                pattern = fr'''(?x)
                (?P<front>{front_barcode}){{e<=1}}
                (?P<inter_user>.{{{space}}})
                (?P<feature>.{{{feature_length}}})
                (?P<inter>.{{{inter_space}}})
                (?P<rear>{rear_barcode}){{e<=1}}
                '''
                regex=re.compile(pattern)
                
            if position == 'pattern_down_spaced':
                down_space_range = space
                inter_space = space_range
                
                pattern = fr'''(?x)
                (?P<front>{front_barcode}){{e<=1}}
                (?P<inter>.{{{inter_space}}})
                (?P<rear>{rear_barcode}){{e<=1}}
                (?P<down_space>.{{{down_space_range}}})
                (?P<feature>.{{{feature_length}}})
                '''
                regex=re.compile(pattern)
                
            # 4. Apply pattern to sequences
            col_name = f"extracted_{pair_name}"
        
            df_copy = df.copy()
            df_copy[col_name] = df_copy['seq'].apply(
                lambda x: regex.search(x).groupdict() if regex.search(x) else None
            )
        
            # 5. Filter and expand extracted results
            df_valid = df_copy[df_copy[col_name].notna()].copy()
            extracted_cols = df_valid[col_name].apply(pd.Series)
            df_combined = pd.concat([df_valid.drop(columns=[col_name]), extracted_cols], axis=1)
        
            # 6. Add metadata info
            df_combined['pair'] = pair_name
            df_combined['front_barcode_name'] = row['front_barcode_name']
            df_combined['rear_barcode_name'] = row['rear_barcode_name']
        
            extracted_dfs.append(df_combined)
        
            # 7. Track counts
            count_records.append({
                'pair': pair_name,
                'front_barcode_name': row['front_barcode_name'],
                'rear_barcode_name': row['rear_barcode_name'],
                'front_barcode_seq': front_barcode,
                'rear_barcode_seq': rear_barcode,
                'count': len(df_combined)
            })
        
        # Final outputs
        df_result = pd.concat(extracted_dfs).reset_index(drop=True) if extracted_dfs else pd.DataFrame()
        df_counts = pd.DataFrame(count_records)
        
        return df_result, df_counts
```

```python
#SRR30861017 is R3_rep2_dTom
ss_ret = pd.read_csv('results/snipseq/facs/SRR30861017_1_ss_assign_ret.csv')
ss_ret_counts = pd.read_csv('results/snipseq/facs/SRR30861017_1_ss_assign_ret_counts.csv')
ss_ret.shape
```

```python
ss_ret_matrix = ss_ret.groupby(['feature', 'pair'], dropna=False).size().reset_index(name='count')
ss_ret_matrix['source'] = 'ss'
ss_ret_matrix['sort'] = 'R3'
ss_ret_matrix['replicate'] = 2
ss_ret_matrix = ss_ret_matrix.rename(columns={'feature':'Barcode','count':'counts'})
ss_ret_matrix['pair'] = ss_ret_matrix['pair'].replace({'dtom': 'dTom','gfp_s': 'GFP','gfp_uns': 'unspliced'})
```

```python
# Keep Barcode as id; melt all other columns
id_cols = ['Barcode']
value_cols = [c for c in kw.columns if c not in id_cols]

kw_long = kw.melt(
    id_vars=id_cols,
    value_vars=value_cols,
    var_name='measurement',
    value_name='counts'
)

# Split 'G1_rep1_GFP' into ['G1', 'rep1', 'GFP'] using simple string ops
parts = kw_long['measurement'].str.split('_', n=2, expand=True)
parts.columns = ['sort', 'rep_token', 'pair']

# Convert rep_token like 'rep1' -> 1 without regex
parts['replicate'] = parts['rep_token'].str.replace('rep', '', regex=False).astype('int64')

# Assemble tidy DataFrame
kw_long = pd.concat([kw_long.drop(columns=['measurement']), parts[['sort', 'replicate', 'pair']]],axis=1)

kw_long = kw_long[['Barcode', 'sort', 'replicate', 'pair', 'counts']]
kw_long['source'] = 'kw'
```

```python
df_r3_rep2 = pd.concat([kw_long[(kw_long.sort=='R3') & (kw_long.replicate==2)], ss_ret_matrix])
```

```python
wide = (df_r3_rep2.pivot_table(index='Barcode', columns='source', values='counts', aggfunc='sum')
          .dropna())  # keep barcodes present in both sources
s1, s2 = wide.columns.tolist()  # e.g., 'kw', 'ss'
```

```python
#if more points above the line means y axis has more counts and vice versa. 
plt.figure(figsize=(3,3))
sns.scatterplot(x=wide[s1], y=wide[s2], alpha=0.6, edgecolor='none')
m = max(wide[s1].max(), wide[s2].max())
plt.plot([0.5, m], [0.5, m], 'r--', lw=1, label='y = x')
plt.xscale('log'); plt.yscale('log')
plt.xlabel(f'{s1} counts'); plt.ylabel(f'{s2} counts')
plt.title('Paired counts per barcode (log–log)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#this show how robust snipseq is. Found less barcodes matched to the unspliced reads. 
# --- Build presence table per (Barcode, pair) across sources ---
present = (df_r3_rep2.assign(present=df_r3_rep2['counts'] > 0)
             .pivot_table(index=['Barcode', 'pair'],
                          columns='source',
                          values='present',
                          aggfunc='max')
             .fillna(False))

# Get source names dynamically (expects exactly 2)
sources = present.columns.tolist()
if len(sources) != 2:
    raise ValueError(f"Expected exactly 2 sources, found: {sources}")
s1, s2 = sources  # e.g., 'kw', 'ss'

# --- Compute coverage per pair ---
def coverage_counts(group):
    only_s1 = (group[s1] & ~group[s2]).sum()
    only_s2 = (group[s2] & ~group[s1]).sum()
    both    = (group[s1] & group[s2]).sum()
    return pd.Series({'Only ' + s1: only_s1,
                      'Both': both,
                      'Only ' + s2: only_s2})

coverage_per_pair = (present.groupby(level='pair')
                              .apply(coverage_counts)
                              .reset_index()
                              .melt(id_vars='pair',
                                    var_name='category',
                                    value_name='barcodes'))

# --- Plot: one panel per pair ---
g = sns.catplot(
    data=coverage_per_pair,
    x='category',
    y='barcodes',
    col='pair',
    kind='bar',
    palette='Set2',
    sharey=False
)
g.set_titles('{col_name}')
g.set_axis_labels('', 'Number of barcodes')
for ax in g.axes.ravel():
    for label in ax.get_xticklabels():
        label.set_rotation(20)
        label.set_horizontalalignment('right')
plt.suptitle('Barcode detection overlap per pair', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Pivot coverage into wide format: rows = pair, cols = category
coverage_wide = coverage_per_pair.pivot(index='pair', columns='category', values='barcodes').fillna(0)

# Ensure consistent category order
cat_order = [f'Only {s1}', 'Both', f'Only {s2}']
coverage_wide = coverage_wide.reindex(columns=cat_order)

# Stacked bar plot
ax = coverage_wide.plot(kind='bar', stacked=True, color=sns.color_palette('Set2', n_colors=3), figsize=(6,3))
ax.set_title('Barcode detection overlap per pair')
ax.set_ylabel('Number of barcodes')
ax.set_xlabel('pair')
ax.legend(title='', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Process and prepare dataframe for downstream analysis and comparison

```python
import glob
srr_map = {
    'SRR30861028': ('G1', 'Rep1'),
    'SRR30861027': ('G2', 'Rep1'),
    'SRR30861026': ('G3', 'Rep1'),
    'SRR30861025': ('R1', 'Rep1'),
    'SRR30861024': ('R2', 'Rep1'),
    'SRR30861023': ('R3', 'Rep1'),
    'SRR30861022': ('G1', 'Rep2'),
    'SRR30861021': ('G2', 'Rep2'),
    'SRR30861020': ('G3', 'Rep2'),
    'SRR30861019': ('R1', 'Rep2'),
    'SRR30861018': ('R2', 'Rep2'),
    'SRR30861017': ('R3', 'Rep2'),
}

def parse_from_filename(path):
    base = os.path.basename(path)               # e.g., 'SRR30861020_1_ss_assign_ret_counts.csv'
    srr = base.split('_', 1)[0]                 # 'SRR30861020'
    sort, replicate = srr_map.get(srr, (None, None))
    return srr, sort, replicate

paths = sorted(glob.glob('results/snipseq/facs/*_1_ss_assign_ret_counts.csv'))

dfs = []
for p in paths:
    srr, sort, replicate = parse_from_filename(p)
    df = pd.read_csv(p)
    df = df.assign(
        filename=p,
        sample=srr,
        sort=sort,
        replicate=replicate,
        source='ss',
    )
    dfs.append(df)

ss_ret_counts = pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()
ss_ret_counts = ss_ret_counts.rename(columns={'count':'counts'})
ss_ret_counts = ss_ret_counts[['sort', 'replicate', 'pair', 'counts', 'source']]
ss_ret_counts['pair'] = ss_ret_counts['pair'].replace({'dtom': 'dTom','gfp_s': 'GFP','gfp_uns': 'unspliced'})
```

```python
#Find barcodes that pass the minimum read count threshold (at least 10 GFP and 10 dTom reads in at least 3 samples)
df = ss_ret_matrix.copy()
# --- Assume `df` is your DataFrame with the columns you listed ---
# Ensure numeric dtype for count columns (coerce non-numeric to NaN -> 0)
count_cols = [c for c in df.columns if c != 'Barcode']
df[count_cols] = df[count_cols].apply(pd.to_numeric, errors='coerce').fillna(0)

# Identify base sample names by stripping the final suffix (_GFP, _dTom, _unspliced)
suffixes = ('_GFP', '_dTom', '_unspliced')
all_samples = sorted({ c.rsplit('_', 1)[0] for c in df.columns if c.endswith(suffixes) })

# Keep only samples that actually have the necessary columns
# (We expect all three, but this is robust to missing ones; unspliced will default to 0 if missing)
valid_samples = []
for s in all_samples:
    has_dtom = f'{s}_dTom' in df.columns
    has_gfp  = f'{s}_GFP' in df.columns
    has_unsp = f'{s}_unspliced' in df.columns
    if has_dtom and (has_gfp or has_unsp):
        valid_samples.append(s)

# Build per-sample GFP and dTom count matrices
gfp_counts = pd.DataFrame(
    {
        s: df.get(f'{s}_GFP', 0).fillna(0).astype(int) + df.get(f'{s}_unspliced', 0).fillna(0).astype(int)
        for s in valid_samples
    },
    index=df.index
)

dtom_counts = pd.DataFrame(
    {
        s: df.get(f'{s}_dTom', 0).fillna(0).astype(int)
        for s in valid_samples
    },
    index=df.index
)

# Apply thresholds: GFP >= 10 AND dTom >= 10 per sample
gfp_ok   = gfp_counts >= 10
dtom_ok  = dtom_counts >= 10
pass_both = gfp_ok & dtom_ok  # True if both thresholds met in that sample

# Count qualifying samples per barcode
min_samples = 3  # "more than 3 samples" => 4. Change to 3 if you meant "at least 3".
pass_count = pass_both.sum(axis=1)

# Filter barcodes
mask_keep = pass_count >= min_samples
filtered_ss = df.loc[mask_keep].copy()

# (Optional) add diagnostics: how many and which samples passed per barcode
samples_meeting = pass_both.apply(lambda row: [s for s, ok in row.items() if ok], axis=1)
filtered_ss['n_samples_meeting_threshold'] = pass_count[mask_keep]
filtered_ss['samples_meeting_threshold'] = samples_meeting[mask_keep].apply(lambda lst: ','.join(lst))

# The resulting barcodes:
kept_barcodes = filtered_ss['Barcode'].tolist()
print(f"Kept {len(kept_barcodes)} barcodes with ≥10 GFP and ≥10 dTom in ≥{min_samples} samples.")
```

```python
# --- Compute coverage per pair ---
def coverage_counts(group):
    only_s1 = (group[s1] & ~group[s2]).sum()
    only_s2 = (group[s2] & ~group[s1]).sum()
    both    = (group[s1] & group[s2]).sum()
    return pd.Series({'Only ' + s1: only_s1,
                      'Both': both,
                      'Only ' + s2: only_s2})
    
present = (df_long.assign(present=df_long['counts'] > 0)
             .pivot_table(index=['Barcode', 'pair'],
                          columns='source',
                          values='present',
                          aggfunc='max')
             .fillna(False))

# Get source names dynamically (expects exactly 2)
sources = present.columns.tolist()
if len(sources) != 2:
    raise ValueError(f"Expected exactly 2 sources, found: {sources}")
s1, s2 = sources  # e.g., 'kw', 'ss'

coverage_per_pair = (present.groupby(level='pair')
                              .apply(coverage_counts)
                              .reset_index()
                              .melt(id_vars='pair',
                                    var_name='category',
                                    value_name='barcodes'))
```

```python
def coverage_counts(group):
    only_s1 = (group[s1] & ~group[s2]).sum()
    only_s2 = (group[s2] & ~group[s1]).sum()
    both    = (group[s1] & group[s2]).sum()
    return pd.Series({'Kowal': only_s1,
                      'Both': both,
                      'Snipseq': only_s2})
    
present_s4 = (df_long_s4.assign(present_s4=df_long_s4['counts'] > 0)
             .pivot_table(index=['Barcode', 'pair'],
                          columns='source',
                          values='present_s4',
                          aggfunc='max')
             .fillna(False))

# Get source names dynamically (expects exactly 2)
sources_s4 = present_s4.columns.tolist()
if len(sources_s4) != 2:
    raise ValueError(f"Expected exactly 2 sources, found: {sources_s4}")
s1, s2 = sources_s4

coverage_per_pair_s4 = (present_s4.groupby(level='pair')
                              .apply(coverage_counts)
                              .reset_index()
                              .melt(id_vars='pair',
                                    var_name='category',
                                    value_name='barcodes'))
```

```python
# Pivot coverage into wide format: rows = pair, cols = category
coverage_wide = coverage_per_pair.pivot(index='pair', columns='category', values='barcodes').fillna(0)

# Ensure consistent category order
cat_order = [f'Only {s1}', 'Both', f'Only {s2}']
coverage_wide = coverage_wide.reindex(columns=cat_order)

x_order = ['dTom', 'GFP', 'GFP_uns']
coverage_wide = coverage_wide.reindex(x_order)

# Define custom colors (match column order)
colors = ['#FFCC00', '#8da0cbff', '#fc8d62ff']  # Only S1, Both, Only S2

# Stacked bar plot
ax = coverage_wide.plot(kind='bar', stacked=True, color=colors, figsize=(6,3),rot=0)
ax.set_title('Overall Barcode Detection')
ax.set_ylabel('Number of barcodes')
ax.set_xlabel('Reporters')
ax.legend(title='', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.savefig('results/snipseq/facs/figures/barcode_detection_per_reporter_stackedbars.png', dpi=300, bbox_inches='tight')
plt.savefig('results/snipseq/facs/figures/barcode_detection_per_reporter_stackedbars.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python

out_dir = 'results/snipseq/facs/figures'
os.makedirs(out_dir, exist_ok=True)

max_radius = 0.28  # largest possible radius for scaling

pairs = present.index.get_level_values('pair').unique()
for p in pairs:
    sub = present.xs(p, level='pair')
    set1 = set(sub.index[sub[s1]])
    set2 = set(sub.index[sub[s2]])

    only1 = len(set1 - set2)
    only2 = len(set2 - set1)
    both  = len(set1 & set2)

    size1 = only1 + both
    size2 = only2 + both

    # scale radii so area reflects counts
    max_size = max(size1, size2, 1)
    r1 = max_radius * np.sqrt(size1 / max_size)
    r2 = max_radius * np.sqrt(size2 / max_size)

    fig, ax = plt.subplots(figsize=(4,3))
    ax.set_aspect('equal')
    ax.axis('off')

    center1 = (0.4, 0.5)
    center2 = (0.6, 0.5)

    circle1 = plt.Circle(center1, r1, color='#ffcc00', alpha=0.8)
    circle2 = plt.Circle(center2, r2, color='#fc8d62ff', alpha=0.8)

    ax.add_patch(circle1)
    ax.add_patch(circle2)

    ax.text(center1[0] - r1*0.5, center1[1], f'{only1}', ha='center', va='center')
    ax.text(center2[0] + r2*0.5, center2[1], f'{only2}', ha='center', va='center')
    ax.text(0.5, 0.5, f'Both\n{both}', ha='center', va='center', fontweight='bold')
    
    # Titles / subtitles
    total = only1 + only2 + both
    pct_both = (both / total * 100) if total > 0 else 0
    ax.set_title(f'Barcode Detection for {p}', fontsize=14)
    ax.text(0.5, 0.1, f'Total unique barcodes across both: {total}  |  Overlap: {pct_both:.1f}%',
            ha='center', fontsize=10)
    
    # Legend-like labels near circles
    ax.text(center1[0], 0.86, 'Kowal', ha='center', fontsize=12, color='#ffcc00')
    ax.text(center2[0], 0.86, 'Snipseq', ha='center', fontsize=12, color='#fc8d62ff')
    
    out_path_pdf = os.path.join(out_dir, f'barcode_detection_per_reporter_venn_{p}.pdf')
    out_path_png = os.path.join(out_dir, f'barcode_detection_per_reporter_venn_{p}.png')
    plt.savefig(out_path_pdf, dpi=300, bbox_inches='tight')
    plt.savefig(out_path_png, dpi=300, bbox_inches='tight')
    plt.show()
    plt.close(fig)
```
